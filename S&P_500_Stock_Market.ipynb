```python
#import necessary libraries
import yfinance as yf
import pandas as pd
import plotly.express as px
```

# Getting S&P500 Stock from yfinance api

```python
#getting historical data for the S&P 500 (^GSPC is the ticker )
sp500 = yf.download('^GSPC', start='2010-01-01', end='2025-04-01')

#reset index so date becomes a column
sp500.reset_index(inplace=True)

#preview the data
sp500.head()
```

```python
sp500.describe() #to see generally what the structure is
```

```python
#dimensions of sp500
sp500.shape
```

```python
#column names to make sure everything looks right before plotting simple trend line
print(sp500.columns)

#first few rows
print(sp500[['Date', 'Close']].head())
```

```python
#flatten the MultiIndex type columns
sp500.columns = [col[0] if col[0] == 'Date' else col[0] + '_' + col[1] for col in sp500.columns]

#reset index if you haven't already
sp500 = sp500.reset_index()
print(sp500.columns)
```

```python
#first we will plot a simple trendline
fig = px.line(
    sp500,
    x='Date',
    y='Close_^GSPC',
    title='S&P 500 Closing Prices (2010â€“2025)',
    labels={'Close_^GSPC': 'Closing Price (USD)', 'Date': 'Date'}
)

fig.show()
```

```python
#lets do a range slider this time to make it interactive!
fig = px.line(
    sp500,
    x='Date',
    y='Close_^GSPC',
    title='S&P 500 Closing Prices (2010â€“2025)',
    labels={'Close_^GSPC': 'Closing Price (USD)', 'Date': 'Date'}
)
fig.update_xaxes(rangeslider_visible=True)  #adds range slider
fig.show()
```
